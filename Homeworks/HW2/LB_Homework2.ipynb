```python
# pandas 
import pandas as pd

# matplotlib
import matplotlib.pyplot as plt

# seaborn
import seaborn as sns

import numpy as np
```

```python
df = pd.read_csv('C:/Users/Lboyl/Documents/PythonCourse/time_series_covid19_confirmed_global.csv')
```

```python
df_italy = df[df['Country/Region'] == 'Italy'] # this is one way to select Italy from the dataframe
df_italy.head()
```

```python
df_italy1 = df_italy.set_index('Country/Region', drop = True) # set pandas dataframe index to country
df_italy2 = df_italy1.drop(columns=['Lat','Long','Province/State'])

data_italy = df_italy2.loc['Italy','2/20/20':'4/5/20'] 
```

```python
plt.plot(data_italy.values,'.-')
plt.xlabel('days')
plt.show()
```

```python
plt.plot(data_italy.values,'.-',color='xkcd:baby poop green')
plt.xlabel('days')
plt.show()
```

```python
plt.plot(data_italy.values,'.-',color='xkcd:baby poop green',linewidth=2)
plt.xlabel('days')
plt.show()
```

```python
plt.plot(data_italy.values,'.-',color='xkcd:baby poop green',linewidth=2)
plt.xlabel('days')
plt.ylabel('Number of Confirmed Cases')
plt.show()
```

```python
plt.plot(data_italy.values,'.-',color='xkcd:baby poop green',linewidth=2)
plt.xlabel('days')
plt.ylabel('Number of Confirmed Cases')
plt.title('Coronavirus Cases in Italy')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20, 10))

plt.plot(data_italy.index,data_italy.values,'.-')
plt.xticks(rotation=90) # rotate the xticks so text is not so tight
plt.xlabel('calendar date')
plt.show()
```

```python

df_us = df[df['Country/Region'] == 'US']
df_us = df_us.set_index('Country/Region', drop = True) # set pandas dataframe index to country
df_us = df_us.drop(columns=['Lat','Long','Province/State'])

data_us = df_us.loc['US','2/28/20':'4/5/20']
```

```python
fig,ax = plt.subplots(1,2,figsize=(15,5)) 

ax[0].plot(data_italy,color='red')
ax[0].set_ylim(0,350000)


ax[1].plot(data_us,color='blue')
ax[1].set_ylim(0,350000)


for a in fig.axes:
    a.tick_params('x',labelrotation=90)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20, 10))

plt.plot(data_italy.index,data_italy.values,'.-',color='red')
plt.plot(data_us.index,data_us.values,'.-',color='blue')
plt.xticks(rotation=90) 
plt.xlabel('calendar date')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20, 10))

log_italy=np.log(data_italy.values)
log_us=np.log(data_us.values)

plt.plot(data_italy.index,log_italy,'.-',color='red')
plt.plot(data_us.index,log_us,'.-',color='blue')
plt.xticks(rotation=90) 
plt.xlabel('calendar date')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
import math

for idxit,itval in zip(data_italy.index,data_italy):
    lgit=math.log(itval)
    plt.plot(idxit,lgit,'.-',color='red')
    
for idxus,usval in zip(data_us.index,data_us):
    lgus=math.log(usval)
    plt.plot(idxus,lgus,'.-',color='blue')
    
plt.xticks(rotation=90) 
plt.xlabel('calendar date')
plt.show()
```

```python
df_brazil = df[df['Country/Region'] == 'Brazil']
df_brazil = df_brazil.set_index('Country/Region', drop = True) # set pandas dataframe index to country
df_brazil = df_brazil.drop(columns=['Lat','Long','Province/State'])
data_brazil = df_brazil.loc['Brazil','2/28/20':'4/5/20']

df_canada = df[df['Country/Region'] == 'Canada']
df_canada = df_canada.set_index('Country/Region', drop = True) # set pandas dataframe index to country
df_canada = df_canada.drop(columns=['Lat','Long','Province/State'])
data_canada = df_canada.loc['Canada','2/28/20':'4/5/20']

df_austria = df[df['Country/Region'] == 'Austria']
df_austria = df_austria.set_index('Country/Region', drop = True) # set pandas dataframe index to country
df_austria = df_austria.drop(columns=['Lat','Long','Province/State'])
data_austria = df_austria.loc['Austria','2/28/20':'4/5/20']

fig, ax = plt.subplots(1,5,figsize=(20, 4))

ax[0].plot(data_italy.index,data_italy,'.-',color='red',label='Italy')
ax[1].plot(data_us.index,data_us,'.-',color='blue',label='USA')
ax[2].plot(data_brazil.index,data_brazil,'.-',color='green',label='Brazil')
ax[3].plot(data_canada.index,data_canada,'.-',color='purple',label='Canada')
ax[4].plot(data_austria.index,data_austria,'.-',color='orange',label='Austria')

for a in ax:
    a.set_xlabel('Date')
    a.set_ylim(0,350000)
    a.set_ylabel('Number of Confirmed Cases')
    a.tick_params('x',labelrotation=90)
    a.legend()
```
